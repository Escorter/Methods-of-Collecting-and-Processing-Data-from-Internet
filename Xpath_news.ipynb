#### Задача
Написать приложение и функцию, которые собирают основные новости с сайта на выбор dzen.ru, lenta.ru, mail.ru . Для парсинга использовать XPath
Структура данных должна содержать:
* название источника
* наименование новости
* ссылку на новость
* дата публикации

```python
# Функция возвращает массив словарей с новостями с lenta.ru.
# Эти словари имеют ключи соотвественно структуре данных в задаче,
# а значения - соответствующий контент

def news_lenta():
    import requests
    from lxml import html
    from pprint import pprint
    
    
    url = 'https://lenta.ru/'
    headers = {
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'
    }
    response = requests.get(url=url, headers=headers)
    dom = html.fromstring(response.text)
    content = dom.xpath("//a[contains(@class, 'card-')]")
    lenta = []
    i=-1
    for item in content:
        
        source = item.xpath(".//*[name()='svg']/*[name()='use']/attribute::*")
        if len(source)==1:
            source = source[0].split('ui-label_')
        if len(source)==2:
            source = str(source[1]) + '.ru'
        else:
            source = 'lenta.ru'
           
        text = item.xpath("//span[contains(@class, 'card-')]/text()")
        text=text[i]
        i+=1
        
        
        link = 'https://lenta.ru/' + item.xpath('./@href')[0]
        
        date = item.xpath('.//time/text()')
        
        news = {'Источник':source, 'Новость':text, 'Ссылка':link, 'Дата':date}
        lenta.append(news)
    return lenta[1::]
        
    
    
```

```python
lentaru = news_lenta()
```
